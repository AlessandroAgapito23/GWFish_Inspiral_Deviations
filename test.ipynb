```python
import GWFish.modules as gw
import numpy as np
import pandas as pd
```

```python
params = {
        "mass_1_source": 1.4,
        "mass_2_source": 1.4,
        "redshift": 0.01,
        "luminosity_distance": 40,
        "theta_jn": 5 / 6 * np.pi,
        "ra": 3.45,
        "dec": -0.41,
        "psi": 1.6,
        "phase": 0,
        "geocent_time": 1187008882,
    }

parameter_values = pd.DataFrame()
for key, item in params.items():
    parameter_values[key] = np.full((1,), item)

fisher_parameters = list(params.keys())
fisher_parameters.remove('redshift')
#fisher_parameters.remove('dec')

network = gw.detection.Network(
    detector_ids=["ET"],
)
```

```python
gw.fishermatrix.analyze_and_save_to_txt(
                                            network = network,
                                            parameter_values = parameter_values,
                                            fisher_parameters=fisher_parameters,
                                            sub_network_ids_list=[[0]], 
                                            waveform_class=gw.waveforms.TaylorF2,
                                            waveform_model='TaylorF2',
                                            population_name='test',
                                            decimal_output_format='%.15f',
                                        )
```
